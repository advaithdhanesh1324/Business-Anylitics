```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/new_retail_data.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Total_Amount'])
plt.title('The TOTAL')
plt.show()
```

```python
print(df.columns)
```

```python
df['Total_Amount'] = df['Total_Amount'].fillna(df['Total_Amount'].mean())
xyg = df['Total_Purchases'].mode()[0]
df['Total_Purchases'] = df['Total_Purchases'].fillna(xyg)
print('Missing values after imputation:')
print(df.isnull().sum())
```

```python
# Feature Engineering
from datetime import datetime
current_year = datetime.now().year
df['Total_Purchases'] = current_year - df['Total_Purchases']
print("DataFrame with new 'Total_Purchases' column:")
print(df.head())
```

```python
numerical_df = df[['Age', 'Total_Purchases', 'Amount', 'Total_Amount', 'Ratings']]
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
